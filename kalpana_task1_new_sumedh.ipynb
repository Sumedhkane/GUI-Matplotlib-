```python
import tkinter as tk
from tkinter.filedialog import askopenfilename
from tkinter.ttk import *
import pandas as pd
import numpy as np
df = pd.DataFrame()
df_numerics_only = pd.DataFrame()
tochoose=[]
a=[]
zipped=[]
n=0
k=0
time=[]


def import_csv_data():
    global v
    csv_file_path = askopenfilename()
    print(csv_file_path)
    v.set(csv_file_path)
    global df
    df=pd.read_csv(csv_file_path)
    global df_numerics_only
    df_numerics_only = df.select_dtypes(include=np.number)
    global tochoose
    tochoose=list(df_numerics_only.columns)
    global a
    a=list(np.arange(0,len(tochoose)))
    global zipped
    zipped= zip(a,tochoose)
    global n
    n=len(tochoose)
    print(df.head())
    


root = tk.Tk()
root.wm_title("ChooseFile")

tk.Label(root, text='File Path').grid(row=0, column=0)
v = tk.StringVar()
entry = tk.Entry(root, textvariable=v).grid(row=0, column=1)
tk.Button(root, text='Browse Data Set',command=import_csv_data).grid(row=1, column=0)
tk.Button(root, text='Close',command=root.destroy).grid(row=1, column=1)

root.mainloop()
```

```python
 # Toplevel object which will
    # be treated as a new window
def viewSelected():
    global k
    k=v.get()
    print(k)
    
 # Toplevel object which will
    # be treated as a new window    
newWindow = tk.Tk()
 
    # sets the title of the
    # Toplevel widget
newWindow.title("Plotting")
 
    # sets the geometry of toplevel
newWindow.geometry("200x200")
 
    # A Label widget to show in toplevel
tk.Label(newWindow,
          text ="Choose your y axis select and close").pack()
     
v=tk.IntVar()    

for val, language in zipped:
        tk.Radiobutton(newWindow, 
                   text=language,
                   padx = 20, 
                   variable=v, value=val,
                   command=viewSelected).pack(anchor=tk.W)
root.mainloop()
 
```
